## boolean型の列を使って色を指定したい

```python
iris['has_large_sepal'] = iris['sepal_length'] + iris['sepal_width'] >= 10
```

```python
# seabornのscatterplot（散布図）を使う

sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='has_large_sepal')
```

```python
sns.pairplot(iris)
```

```python
# あれ、これでエラーになったのを見たんだけど。バージョンの違いによる影響かなぁ……

sns.pairplot(iris, hue='has_large_sepal')
```
